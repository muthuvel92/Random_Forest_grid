```python
#importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read File from Csv file using pandas
dataset = pd.read_csv('insurance_pre.csv')
```

```python
#Viewing imported dataset
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset
```

```python
#Viewing dataset Columns
dataset.columns
```

```python
#Assignment of independant variable (Inputs)
independant = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependant = dataset[['charges']]
```

```python
#spliting train and test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independant,dependant, test_size= 0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#Parameter grid is the dictionary keys are names of parameters and values are list of settings to try
#Import GridSearch CV, then required model and enclose respective model parameters in param_grid
#assign variable to gridsearch parameters
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
              'n_estimators':[10,100],'max_features':['auto','sqrt','log2']}
grid = GridSearchCV(RandomForestRegressor(),param_grid, refit = True,verbose = 3,n_jobs = -1)
#fitting model for grid search
grid.fit(X_train,y_train)
```

```python
#print best parameter after turing
#grid.best_params_: This gives you the best parameters found during the grid search.
re = grid.cv_results_
print("R_score value for best parameter {}:".format(grid.best_params_))
```

```python
#Creating Dataframe from dictionary
table = pd.DataFrame.from_dict(re)
table
```

```python
#Assign grid.predict from input lists
Future_Prediction = grid.predict([[age_input,bmi_input,child_input,smoker_input,sex_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
